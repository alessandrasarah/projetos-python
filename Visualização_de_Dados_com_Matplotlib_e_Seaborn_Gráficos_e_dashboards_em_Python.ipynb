```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# definir os dados a serem utilizados
x_line = np.linspace(0, 10, 100)
y_line = np.sin(x_line)
```

```python
# gerar o gráfico de linhas ou plot

fig = plt.figure(figsize=(20, 10))
fig.set_facecolor('#B3E5FC')
plt.plot(x_line, y_line)
plt.xlabel('Rótulo do eixo X')
plt.ylabel('Rótulo do eixo Y')
plt.title('Título do gráfico')
plt.show()
```

Subplot ou Subgráficos


```python
plt.subplots(2,2)
```

```python
fig, axs=plt.subplots(2,2, figsize=(10,5), layout = 'constrained')

#Cores


fig.set_facecolor('#B3E5FC')
axs[0,0].set_facecolor('#AEC6CF')
axs[0,1].set_facecolor('#FFFFBA')
axs[1,0].set_facecolor('#FFD1DC')
axs[1,1].set_facecolor('#BFFCC6')

#Título
axs[0,0].set_title('Título 1')
axs[0,1].set_title('Título 2')
axs[1,0].set_title('Título 3')
axs[1,1].set_title('Título 4')
```

```python
#Gráfico de barras ou Bar

#definindo os dados a serem utilizados

X = ['Chandler','Joey','Monica','Phoebe']
y =[26,10,15,24]
```

```python
plt.style.use('ggplot')

fig, axs=plt.subplots(figsize=(10,5), layout = 'constrained')
axs.set_xlabel('Rótulo do eixo X', fontsize = 9)
axs.set_ylabel('Rótulo do eixo Y', fontsize = 9)
axs.set_title('Título do gráfico', fontsize = 14)
plt.bar(X,y)
plt.show()
```

Seaborn


```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# definindo dados

df =sns.load_dataset('tips')
df.head()
```

```python
# total de gorjetas por dia

sns.barplot(x='day', y='tip', data=df)
```

```python
# relação entre o valor da gorjeta e o valor da conta

sns.scatterplot(x='tip', y='total_bill', data=df)
plt.title('Valor da Gorjeta x Valor da Conta')
plt.show()
```

```python
# Comparação de gorjetas por gênero

sns.set_style('dark')

sns.boxplot(x='sex', y='tip', data=df)
plt.title('Gorjeta por Gênero')
plt.show()
```
